**One place to keep lightweight, safe feature engineering (timestamps, groupings, bands).**

```python
# Cell 3: Helpers (normalize + safe feature engineering)
def normalize_cols(df: pd.DataFrame) -> pd.DataFrame:
    out = df.copy()
    out.columns = [c.strip().upper().replace(" ", "_") for c in df.columns]
    return out

def engineer_features(df: pd.DataFrame) -> pd.DataFrame:
    d = df.copy()

    # SUBMITTED-derived (safe, robust to bad timestamps)
    if "SUBMITTED" in d.columns:
        d["SUBMITTED"] = pd.to_datetime(d["SUBMITTED"], errors="coerce")
        d["SUBMITTED_HOUR"]  = d["SUBMITTED"].dt.hour
        d["SUBMITTED_DOW"]   = d["SUBMITTED"].dt.dayofweek
        d["SUBMITTED_MONTH"] = d["SUBMITTED"].dt.month
        d["WEEKEND_FLAG"]    = d["SUBMITTED"].dt.dayofweek.isin([5,6]).astype("float")

        def hour_bucket(h):
            if pd.isna(h): return "Unknown"
            h = int(h)
            if 5 <= h < 12:  return "morning"
            if 12 <= h < 17: return "afternoon"
            if 17 <= h < 22: return "evening"
            return "night"
        d["SUBMITTED_PERIOD"] = d["SUBMITTED"].dt.hour.map(hour_bucket)
        d["MONTH_IN_QUARTER"] = ((d["SUBMITTED"].dt.month - 1) % 3 + 1).astype("float")

    # Income bands (narrow income is already in 000s)
    def income_band(v):
        if pd.isna(v): return "unknown"
        v = int(v)
        if v < 50: return "low"
        if v < 100: return "mid"
        if v < 150: return "upper"
        return "high"

    if "ESTIMATED_INCOME" in d.columns:
        d["ESTIMATED_INCOME"] = pd.to_numeric(d["ESTIMATED_INCOME"], errors="coerce")
        d["EST_INC_BAND"] = d["ESTIMATED_INCOME"].map(income_band)

    if "MEDIAN_FAMILY_HOUSEHOLD_INCOME" in d.columns:
        mfi = pd.to_numeric(d["MEDIAN_FAMILY_HOUSEHOLD_INCOME"], errors="coerce")
        d["MFI_BAND"] = pd.qcut(mfi, q=4, labels=["q1_low","q2_midlow","q3_midhigh","q4_high"]).astype(str)

    # Source grouping
    def source_group(s):
        if pd.isna(s): return "unknown"
        s = str(s).lower()
        if s in ("g","google"): return "google"
        if s in ("fb","facebook"): return "facebook"
        if s in ("s","search","search_partner"): return "search_partner"
        if s in ("it","iterable"): return "email_iterable"
        return s
    if "SOURCE" in d.columns:
        d["SOURCE_GROUP"] = d["SOURCE"].map(source_group)

    # Keyword grouping (very light heuristic)
    if "KEYWORD" in d.columns:
        def kw_group(k):
            if pd.isna(k): return "other"
            k = str(k).lower()
            if "life alert" in k or "lifealert" in k: return "life_alert"
            if "medical" in k and "alert" in k: return "medical_alert"
            if "cost" in k or "price" in k: return "pricing"
            if "fallen" in k: return "fall_help"
            return "other"
        d["KEYWORD_GROUP"] = d["KEYWORD"].map(kw_group)

    return d
```

**Cell 4: Robust Dataset Preparation**

In this step, we clean and prepare the dataset so that all later analysis is reliable and reproducible.  
- We **normalize column names** and safely engineer time-based features.  
- The target variable **SOLD** is confirmed and coerced to 0/1.  
- Likely numeric columns (like age, income, household size) are coerced to numbers and lightly **capped at realistic ranges** so extreme outliers don’t destabilize the model.  
- We add **missingness flags** (e.g., `AGE_MISSING`) so the model can learn if “not reported” itself carries signal.  
- Very large categorical features (CITY, KEYWORD, PLACEMENT) are **rare-capped** so tiny groups don’t overfit.  

This ensures the dataset is business-ready, statistically robust, and reproducible before moving into EDA and modeling.

```python
# Cell 4: Robust dataset prep BEFORE EDA/Modeling
"""
Purpose :
- Kept all dataset hygiene in one place so the notebook is easy to re-run and audit.
- Normalized column names, engineered safe time features, and confirmed the target.

- Coerce numeric-type candidates, add light caps to avoid outliers, and add _MISSING flags
  (the model can learn that 'was missing' is sometimes might informative).
- Normalized common Y/N fields to 0/1 so plots and models are consistent.

- Rare-cap - high-cardinality categoricals (CITY/KEYWORD/PLACEMENT) to reduce overfitting.
- Build `features_df`, which is the single source of truth used downstream.

Note:
- We DO NOT impute here. All imputation stays inside the sklearn Pipeline (fit on train only),
  which avoids leakage and keeps things reproducible.
"""

from pandas.api.types import is_numeric_dtype
import numpy as np, pandas as pd
import seaborn as sns, matplotlib.pyplot as plt

# 1) Normalize + engineer features
leads = normalize_cols(leads_raw)
leads = engineer_features(leads)

# 2) Target confirmation (predict SOLD = 1)
TARGET = "SOLD"
assert TARGET in leads.columns, "SOLD target not found."
leads[TARGET] = pd.to_numeric(leads[TARGET], errors="coerce").clip(0, 1).fillna(0).astype(int)

# 3) Coerce common numeric candidates (robust to stringy inputs)
num_candidates = [
    "AGE","NUM_PERSONS","NUM_CHILDREN","NUM_ADULTS",
    "ESTIMATED_INCOME","LENGTH_OF_RESIDENCE","MEDIAN_FAMILY_HOUSEHOLD_INCOME"
]
for col in num_candidates:
    if col in leads.columns:
        leads[col] = pd.to_numeric(leads[col], errors="coerce")

# Detect numeric columns (excluding target) after coercion
numeric_cols = [c for c in leads.columns if c != TARGET and is_numeric_dtype(leads[c])]

# 4) Light winsorization (stable caps; avoids distorting signal)
def cap_to_p99(s: pd.Series):
    if s.dropna().empty:
        return s
    p99 = s.quantile(0.99)
    return s.clip(upper=p99)

if "AGE" in leads.columns:
    leads["AGE"] = leads["AGE"].clip(lower=18, upper=100)
if "ESTIMATED_INCOME" in leads.columns:
    leads["ESTIMATED_INCOME"] = leads["ESTIMATED_INCOME"].clip(upper=500)  # band code upper bound
if "MEDIAN_FAMILY_HOUSEHOLD_INCOME" in leads.columns:
    leads["MEDIAN_FAMILY_HOUSEHOLD_INCOME"] = cap_to_p99(leads["MEDIAN_FAMILY_HOUSEHOLD_INCOME"])
if "LENGTH_OF_RESIDENCE" in leads.columns:
    p99_len = leads["LENGTH_OF_RESIDENCE"].quantile(0.99)
    leads["LENGTH_OF_RESIDENCE"] = leads["LENGTH_OF_RESIDENCE"].clip(upper=min(60, p99_len))
for hh_col, hard_cap in [("NUM_CHILDREN", 10), ("NUM_ADULTS", 10), ("NUM_PERSONS", 12)]:
    if hh_col in leads.columns:
        p99 = leads[hh_col].quantile(0.99)
        leads[hh_col] = leads[hh_col].clip(upper=min(hard_cap, p99))

# 5) Numeric missingness flags ONLY (no imputation here)
for col in numeric_cols:
    if leads[col].isna().any():
        leads[f"{col}_MISSING"] = leads[col].isna().astype("int8")

# 6) Binary normalization (Y/N → 1/0) BEFORE we build features_df
binary_map = {"YES": 1, "Y": 1, "1": 1, "TRUE": 1,
              "NO": 0,  "N": 0, "0": 0, "FALSE": 0}
binary_cols = [
    "AUTOMATIC_FALL_DETECTION", "INPUT_ADDRESS_MATCHES", "INPUT_PHONE_MATCHES",
    "INPUT_EMAIL_MATCHES", "PRESENCE_OF_CREDIT_CARD"
]
for col in binary_cols:
    if col in leads.columns:
        leads[col] = (
            leads[col]
            .astype(str).str.strip().str.upper()
            .replace({"NAN": None})
            .map(binary_map)
            .astype("float")
        )

# 7) Rare-cap very high-cardinality categoricals
def rare_cap(series: pd.Series, min_count: int = 50, other_label: str = "Other"):
    vc = series.value_counts(dropna=False)
    keep = vc[vc >= min_count].index
    return series.where(series.isin(keep), other_label)

high_card_cols = [c for c in ["CITY", "KEYWORD", "PLACEMENT"] if c in leads.columns]
for col in high_card_cols:
    leads[col] = leads[col].astype("object").where(leads[col].notna(), "Unknown").astype(str).str.strip()
    leads[col] = rare_cap(leads[col], min_count=50, other_label="Other")

# 8) Build the working feature frame (drop raw IDs/time)
drop_cols = [c for c in ["UNIQUE_ID", "SUBMITTED"] if c in leads.columns]
features_df = leads.drop(columns=drop_cols, errors="ignore")

# 9) Sanity prints
print("Numeric columns detected:", len(numeric_cols))
print("Missing flags added:", [c for c in leads.columns if c.endswith("_MISSING")])
print("Rare-capped columns:", high_card_cols or "(none)")
print(f"Rows: {features_df.shape[0]:,} | Columns (incl. SOLD): {features_df.shape[1]:,}")

print("Target column:", TARGET)
print(features_df[TARGET].value_counts(normalize=True).round(3))
conv_rate = features_df[TARGET].mean()
print(f"Overall conversion rate: {conv_rate:.3%}")

sns.countplot(x=features_df[TARGET])
plt.title("Distribution of Target (SOLD)")
plt.xlabel("SOLD (0=Not Converted, 1=Converted)")
plt.ylabel("Count")
plt.tight_layout(); plt.show()
```

This cell investigates interaction effects between two categorical variables (e.g., Urgency × Source, Urgency × Device, Network × Urgency). Instead of looking at each variable in isolation, we analyze how their combinations influence conversion rates.
- Conversion heatmaps highlight where conversion rates are higher or lower across intersections.
- A corresponding sample count heatmap ensures reliability by filtering out low-volume combinations (to avoid noisy or misleading insights).

**This approach helps us identify targetable segments that might be overlooked in single-variable analysis.**

```python
# Cell: Interaction heatmap
"""
reveal combinations that lift conversion (e.g., Urgency × Source).
filtered low-volume cells to avoid noisy rates.
"""
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def conv_heatmap(df, row_feat, col_feat, target="SOLD",
                 min_count=50, top_rows=None, top_cols=None):
    d = df[[row_feat, col_feat, target]].copy()
    d[row_feat] = d[row_feat].astype("object").fillna("Unknown")
    d[col_feat] = d[col_feat].astype("object").fillna("Unknown")

    # optionally limit to top by volume to keep readable
    if top_rows:
        keep_r = d[row_feat].value_counts().head(top_rows).index
        d = d[d[row_feat].isin(keep_r)]
    if top_cols:
        keep_c = d[col_feat].value_counts().head(top_cols).index
        d = d[d[col_feat].isin(keep_c)]

    agg = d.groupby([row_feat, col_feat])[target].agg(['count','mean']).reset_index()
    agg = agg[agg['count'] >= min_count]
    if agg.empty:
        print(f"[{row_feat} × {col_feat}] no cells with count >= {min_count}")
        return

    rate = agg.pivot(index=row_feat, columns=col_feat, values='mean')
    cnt  = agg.pivot(index=row_feat, columns=col_feat, values='count')

    plt.figure(figsize=(8,5))
    sns.heatmap(rate, annot=True, fmt=".02f", cmap="Blues", cbar_kws={"label":"Conversion Rate"})
    plt.title(f"Conversion Rate by {row_feat} × {col_feat} (min_count={min_count})")
    plt.xlabel(col_feat); plt.ylabel(row_feat); plt.tight_layout(); plt.show()

    plt.figure(figsize=(8,5))
    sns.heatmap(cnt, annot=True, fmt=".0f", cmap="Greys", cbar_kws={"label":"Sample Count"})
    plt.title(f"Sample Count by {row_feat} × {col_feat}")
    plt.xlabel(col_feat); plt.ylabel(row_feat); plt.tight_layout(); plt.show()

# Examples (use what you have):
pairs = []
if "URGENCY" in leads.columns and "SOURCE_GROUP" in leads.columns:
    pairs.append(("URGENCY","SOURCE_GROUP"))
if "URGENCY" in leads.columns and "DEVICE" in leads.columns:
    pairs.append(("URGENCY","DEVICE"))
if "NETWORK" in leads.columns and "URGENCY" in leads.columns:
    pairs.append(("NETWORK","URGENCY"))

for r,c in pairs:
    print(f"=== Heatmaps: {r} × {c} ===")
    conv_heatmap(leads, r, c, target=TARGET, min_count=50, top_rows=10, top_cols=8)
```

```python
#  Conversion tables with 95% confidence intervals
"""
report conversion rate with uncertainty → more credible comparisons.

Uses Wilson interval for proportions.

"""
from statsmodels.stats.proportion import proportion_confint

def conv_table_with_ci(df, feat, target="SOLD"):
    t = df.groupby(feat)[target].agg(['sum','count']).rename(columns={'sum':'conversions'})
    t['rate'] = t['conversions'] / t['count']
    lo, hi = proportion_confint(t['conversions'], t['count'], alpha=0.05, method='wilson')
    t['ci_low'] = lo; t['ci_high'] = hi
    return t.sort_values('rate', ascending=False)

# Example categories (adjust as needed)
for feat in [f for f in ["URGENCY","SOURCE_GROUP","DEVICE","NETWORK","STATE"] if f in leads.columns]:
    print(f"\n== {feat}: conversion rate with 95% CI ==")
    display(conv_table_with_ci(leads, feat, target=TARGET).head(15))
```

### Cell 8: Preprocessing Pipeline (Impute + Scale + Encode)

Here we prepare the data in a way that mimics how it would work in production:

- **Numeric fields** → fill missing values with the **median**, then scale them (so models like Logistic Regression stay stable).  
- **Categorical fields** → fill blanks with the **most common category**, then convert to machine-readable dummies (One-Hot Encoding{handle_unknown="ignore") }).  
- The pipeline **learns imputation values only from the training data**, then applies them to the validation set.  
  - This avoids “peeking” at validation data (no leakage).  
  - It matches how the model would behave in the real world when scoring unseen leads.  

This setup ensures a **reproducible, production-ready workflow**: clean, consistent inputs no matter how messy new data arrives.

```python
# Cell 8: Preprocessor (impute + scale + OHE)
coded_cats = set([
    "FIRST_NAME_MATCH","LAST_NAME_MATCH","INPUT_ADDRESS_MATCHES","INPUT_PHONE_MATCHES","INPUT_EMAIL_MATCHES",
    "PERSON_TYPE","OCCUPATION_GROUP","PHONE_ACTIVE","PHONE_PAST_2_MONTHS_PHONE_USAGE","PHONE_PAST_12_MONTHS_PHONE_USAGE",
    "PRESENCE_OF_CHILD","PRESENCE_OF_CREDIT_CARD","BUYING_POWER_SCORE","NET_ASSET_VALUE","MATCH_TYPE","DEVICE",
    "NETWORK","MEDIUM","SOURCE","SOURCE_GROUP","PLACEMENT","KEYWORD","KEYWORD_GROUP",
    "WIT_SHOPPINGFORWHO","AUTOMATIC_FALL_DETECTION","DEVICE_USEDWHERE","LEAVE_HOME","URGENCY","STATE","CITY",
    "SUBMITTED_PERIOD","EST_INC_BAND","MFI_BAND"
])

def split_feature_types(df):
    num_feats, cat_feats = [], []
    for c in df.columns:
        s_num = pd.to_numeric(df[c], errors="coerce")
        if c in coded_cats:
            cat_feats.append(c)
        elif s_num.notna().sum() > 0 and s_num.dtype.kind in "if":
            num_feats.append(c)
        else:
            cat_feats.append(c)
    return num_feats, cat_feats

num_features, cat_features = split_feature_types(X_train)

preprocessor = ColumnTransformer(
    transformers=[
        ("num", Pipeline(steps=[
            ("imputer", SimpleImputer(strategy="median")),
            ("scale",   StandardScaler())
        ]), num_features),
        ("cat", Pipeline(steps=[
            ("imputer", SimpleImputer(strategy="most_frequent")),
            ("ohe",     OneHotEncoder(handle_unknown="ignore"))
        ]), cat_features),
    ],
    remainder="drop"
)
print("num_features:", len(num_features), "| cat_features:", len(cat_features))
```

**Modular, returns metrics and probabilities so later cells can reuse them.**

This helper function is like a “report card” for our models.

•It calculates key performance metrics (accuracy, precision, recall, ROC AUC, PR AUC, log-loss) so we can judge how well the model is separating converters from non-converters.

•It shows diagnostic plots (ROC curve, Precision-Recall curve, and calibration curve) to visually check if the probabilities the model gives make sense.

• It also looks at the top 10% of leads by model score and compares their conversion rate to the base rate, so we can see how much lift the model provides in prioritizing high-quality leads.

```python
# Cell 10: Evaluate helper
def evaluate(model, Xv, yv, label="Model", top_quantile=0.9):
    proba = model.predict_proba(Xv)[:,1]
    pred  = (proba >= 0.5).astype(int)

    metrics = {
        "roc_auc": roc_auc_score(yv, proba),
        "pr_auc":  average_precision_score(yv, proba),
        "log_loss": log_loss(yv, proba),
        "acc":      accuracy_score(yv, pred),
        "precision": precision_score(yv, pred, zero_division=0),
        "recall":    recall_score(yv, pred, zero_division=0)
    }
    print(f"{label} — ROC AUC: {metrics['roc_auc']:.4f} | PR AUC: {metrics['pr_auc']:.4f} | Log-loss: {metrics['log_loss']:.4f} | Acc: {metrics['acc']:.3f} | P/R: {metrics['precision']:.3f}/{metrics['recall']:.3f}")
    print("\nClassification report @0.5 threshold:\n")
    print(classification_report(yv, pred, digits=3))

    # Curves
    RocCurveDisplay.from_predictions(yv, proba); plt.title(f"{label} — ROC"); plt.show()
    PrecisionRecallDisplay.from_predictions(yv, proba); plt.title(f"{label} — PR"); plt.show()

    # Lift at top quantile
    thr = np.quantile(proba, top_quantile)
    mask = proba >= thr
    base = yv.mean()
    top  = yv[mask].mean()
    lift = (top / base) if base > 0 else np.nan
    print(f"Top {int((1-top_quantile)*100)}% threshold: {thr:.4f} | Base CR: {base:.4f} | Top-bucket CR: {top:.4f} | Lift: {lift:.2f}x")

    # Calibration
    frac_pos, mean_pred = calibration_curve(yv, proba, n_bins=10, strategy="quantile")
    plt.figure(); plt.plot(mean_pred, frac_pos, marker="o"); plt.plot([0,1],[0,1],"--", color="gray")
    plt.title(f"{label} — Calibration"); plt.xlabel("Mean predicted"); plt.ylabel("Fraction positive"); plt.tight_layout(); plt.show()

    return metrics, proba
```

```python
# Cell 17: Threshold sweep (Logistic & XGB)
from sklearn.metrics import confusion_matrix

def threshold_sweep(y_true, y_proba, thresholds=None):
    if thresholds is None:
        thresholds = np.linspace(0.05, 0.95, 19)
    rows = []
    for t in thresholds:
        pred = (y_proba >= t).astype(int)
        tn, fp, fn, tp = confusion_matrix(y_true, pred).ravel()
        prec = tp / (tp + fp) if (tp + fp) else 0.0
        rec  = tp / (tp + fn) if (tp + fn) else 0.0
        f1   = 2*prec*rec/(prec+rec) if (prec+rec) else 0.0
        rows.append({"threshold": t, "precision": prec, "recall": rec, "f1": f1, "selected_rate": pred.mean()})
    return pd.DataFrame(rows)

sw_log = threshold_sweep(y_val, proba_log_tuned)
sw_xgb = threshold_sweep(y_val, proba_xgb_tuned)

print("Logistic — best F1:")
display(sw_log.sort_values("f1", ascending=False).head(5))

print("XGB — best F1:")
display(sw_xgb.sort_values("f1", ascending=False).head(5))

print("Logistic — best precision with recall ≥ 0.60:")
display(sw_log[sw_log["recall"] >= 0.60].sort_values("precision", ascending=False).head(3))
```

```python
#  Cell 18A (fixed): Per-lead probability scores (Logistic & XGB)
# Returns a probability for EVERY validation row, with business context columns attached.
# Also writes a CSV you can hand to stakeholders.

# 1) Get probabilities from tuned models
proba_log_all = best_logit.predict_proba(X_val)[:, 1]
proba_xgb_all = best_xgb.predict_proba(X_val)[:, 1]

# 2) Choose a few readable business columns (keep only those that exist)
id_like_cols = [c for c in ["UNIQUE_ID"] if c in leads.columns]
context_cols = [c for c in ["URGENCY","SOURCE_GROUP","NETWORK","DEVICE","STATE","CITY"] if c in leads.columns]
present_cols = id_like_cols + context_cols

# 3) Recover the original (unencoded) validation slice from *leads* (not features_df)
# Indices align because features_df was derived from leads without reindexing.
if present_cols:
    val_context = leads.loc[X_val.index, present_cols].copy()
else:
    val_context = pd.DataFrame(index=X_val.index)

# 4) Build the scored table
scored_val = (
    pd.DataFrame({
        "pred_prob_logit": proba_log_all,
        "pred_prob_xgb":   proba_xgb_all,
        "actual":          y_val.values
    }, index=X_val.index)
    .join(val_context, how="left")
)

# 5) Quick peek: top & bottom by XGB score
print("Top 10 leads by XGB probability:")
display(scored_val.sort_values("pred_prob_xgb", ascending=False).head(10))

print("Bottom 10 leads by XGB probability:")
display(scored_val.sort_values("pred_prob_xgb", ascending=True).head(10))

# 6) (Optional) export a review file
OUT_PATH = "validation_scored.csv"
scored_val.to_csv(OUT_PATH, index=False)
print(f"Saved per-lead scores to: {OUT_PATH}")

# 7) (Optional) quick decile lift table
bins = pd.qcut(scored_val["pred_prob_xgb"], q=10, duplicates="drop")
lift_table = scored_val.groupby(bins)["actual"].agg(['mean','size']).rename(columns={"mean":"conversion_rate","size":"n"}).reset_index()
display(lift_table.sort_values("conversion_rate", ascending=False))
```

```python
# ---- Cell 19: Final Model Comparison Summary ----
summary_rows = []

def add_summary(name, metrics):
    summary_rows.append({
        "Model": name,
        "ROC AUC": round(metrics["roc_auc"], 4),
        "PR AUC": round(metrics["pr_auc"], 4),
        "Log-loss": round(metrics["log_loss"], 4),
        "Accuracy": round(metrics["acc"], 3),
        "Precision": round(metrics["precision"], 3),
        "Recall": round(metrics["recall"], 3)
    })

# Collect metrics
add_summary("Dummy (majority)", {"roc_auc":0.5, "pr_auc":y_val.mean(), "log_loss":1.0,
                                 "acc":(y_val==0).mean(), "precision":0, "recall":0})
add_summary("Logistic (baseline)", metrics_log_base)
add_summary("Logistic (tuned)",    metrics_log_tuned)
if "metrics_log_smote" in globals():
    add_summary("Logistic + SMOTE",   metrics_log_smote)
add_summary("XGB (baseline)",     metrics_xgb_base)
add_summary("XGB (tuned)",        metrics_xgb_tuned)

summary_df = pd.DataFrame(summary_rows)
display(summary_df.sort_values("ROC AUC", ascending=False))
```

## Conclusion

- Both models (Logistic and XGBoost) achieved **~0.68 ROC AUC and ~0.11 PR AUC**, well above the dummy baseline.  
- **XGBoost (tuned)** slightly outperformed Logistic in ROC/PR AUC, but Logistic remains more interpretable.  
- **Calibration** improved Logistic probability estimates significantly (Brier score ↓).  
- Threshold analysis shows we can tune for higher recall (important for lead capture) at the cost of precision.  
- Recommendation:  
  - Deploy **Logistic (tuned + calibrated)** if transparency is prioritized.  
  - Deploy **XGB (tuned)** if marginal lift in predictive power is more valuable than interpretability.
